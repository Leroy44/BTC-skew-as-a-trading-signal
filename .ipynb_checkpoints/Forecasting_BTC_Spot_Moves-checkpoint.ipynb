```python
import json
import pandas as pd
import datetime
from pathlib import Path
import hvplot.pandas
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
# Import csv files from github

url = 'https://raw.githubusercontent.com/Leroy44/BTC-skew-as-a-trading-signal/main/Data/all_data.csv'
all_data_df = pd.read_csv(url)
```

```python
# Define path, datetime format, parse dates

all_data_df = pd.read_csv(url, index_col = "date", infer_datetime_format= True, parse_dates=True)
```

```python
# Mapping and plotting full bitcoin price dataframe - hourly intervals
btc_close = pd.DataFrame(all_data_df["btc_close"])

btc_close.plot(figsize=(16,8))
plt.ylabel('btc price')
plt.title('btc price', fontsize=18)
plt.legend(fontsize=10)
```

```python
# Slice data set and interpolate between missing values
chopped_data_df = all_data_df.loc["2020-03-01 18:00:00":"2021-12-31 19:00:00"]
chopped_data_int_df = chopped_data_df.interpolate('linear')
chopped_data_int_df.head(5)
```

```python
# Create new variable for % change between intervals

# Spot prices
chopped_data_int_df["btc_close_pct"] = chopped_data_df['btc_close'].pct_change()
chopped_data_int_df["eth_close_pct"] = chopped_data_df['eth_close'].pct_change()
chopped_data_int_df["dxy_close_pct"] = chopped_data_df['dxy_close'].pct_change()
chopped_data_int_df["es_close_pct"] = chopped_data_df['es_close'].pct_change()
chopped_data_int_df["gc_close_pct"] = chopped_data_df['gc_close'].pct_change()
chopped_data_int_df["us_close_pct"] = chopped_data_df['us_close'].pct_change()


# Skew values
#btc_thirtyFiveDelta7DayExp
#btc_twentyFiveDelta7DayExp
#btc_fifteenDelta7DayExp
#btc_fiveDelta7DayExp
#btc_thirtyFiveDelta7DayExp
#btc_thirtyFiveDelta7DayExp
```

```python
# sample plot of vol data
btc_twentyFiveDelta30DayExp = pd.DataFrame(all_data_df["btc_twentyFiveDelta30DayExp"])

btc_twentyFiveDelta30DayExp.plot(figsize=(16,8))
plt.ylabel('btc 25d skew')
plt.title('btc 25-delta skew - 30Day', fontsize=18)
plt.legend(fontsize=10)
```

```python
cleaned_data_df = chopped_data_int_df.drop(["btc_close","eth_close","dxy_close","gc_close","us_close","es_close"], axis=1)
```

```python
cleaned_data_df.head()
```

```python
#remove dependent variable
X=cleaned_data_df.drop(["btc_close_pct"],axis=1)
```

```python
#shift dependent variable back to line up forward return
z=cleaned_data_df["btc_close_pct"].shift(-1)
```

```python
X.head()
```

```python
z.head()
```

```python
z.std()
```

```python
z.hist()
```

```python
#Creating Classification
y_list=[]
for i in range(0,len(z)):
  if z[i] < -0.03: 
    y_list.append(-3),
  elif (z[i] <-0.02 and z[i] >= -0.03):
    y_list.append(-2),
  elif (z[i] <-0.01 and z[i] >= -0.02):
    y_list.append(-1),
  elif (z[i] <= 0.02 and z[i] > 0.01):
    y_list.append(1),
  elif (z[i] <= 0.03 and z[i] > 0.02):
    y_list.append(2),
  elif z[i] > 0.03: 
    y_list.append(3)
  else:
    y_list.append(0)
```

```python
#converting the y list into a series with the relevant date-time index
y = pd.Series(y_list,index=z.index)
```

```python
y.value_counts()
```

```python
len(y)
```

```python
#calculating 60% for training, 20% testing, 20% validation
training_end = round(len(y) * .6)
test_end = round(len(y)*.8)
display(training_end)
display(test_end)
```

```python
#creating training sets
X_train = X.iloc[1:training_end,]
y_train = y.iloc[1:training_end,]
```

```python
#creating testing sets
X_test=X.iloc[training_end:test_end,]
y_test=y.iloc[training_end:test_end,]
```

```python
#creating validation sets
X_validate=X.iloc[test_end:,]
y_validate=y.iloc[test_end:,]
```

```python
X_train.head()
```

```python
#scale the data
scaler = StandardScaler()

#apply scaler model to fit the X-train data
X_scaler = scaler.fit(X_train)

#Transform X_train, X_test, X_validate using X_scaler
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
X_validate_scaled = X_scaler.transform(X_validate)
```

```python
#calc number of columns
X_train.shape
```

```python
#LSTM model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

model = Sequential()
model.add(LSTM(
    units=61, return_sequences=True,
    input_shape=(X_train.shape[1], 1)))
model.add(Dropout(0.2))
model.add(LSTM(units=61, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=61))
model.add(Dropout(0.2))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.summary()
```

```python
#train the model
model.fit(X_train_scaled, y_train, epochs=2, shuffle=False, batch_size=1, verbose=1)
```

```python
#calculate predicted values using X_test
predicted_y_test =  model.predict(X_test)

```

```python
predictions_df = pd.DataFrame(index=X_test.index)
predictions_df["Predicted"] = predicted_y_test
predictions_df["Actual"] = y_test
predictions_df.head(25)
```

```python
predictions_df["Predicted"].hist()
```

```python
from sklearn.metrics import classification_report
#create classification report
btc_testing_report = classification_report(y_test, predicted_y_test)
print(btc_testing_report)
```
